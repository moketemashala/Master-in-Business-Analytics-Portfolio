#### Housing behavior:

The housing industry can experience exponential behavior due to several factors:

1. Population Growth: As the population grows, the demand for housing increases. This can lead to an exponential increase in the demand for housing in certain areas, especially in cities with limited land for expansion.
2. Economic Growth: Economic growth can also drive an increase in demand for housing. As people's incomes increase, they may be able to afford better quality housing or move to more expensive areas, leading to an increase in demand for higher-end housing.
3. Interest Rates: Low-interest rates can also fuel demand for housing. When interest rates are low, it becomes more affordable for people to buy homes, which can lead to an increase in demand for housing.
4. Speculation: Speculation can also contribute to exponential behavior in the housing industry. When people believe that housing prices will continue to rise, they may be more likely to invest in the market, driving up demand and prices further. However, it's worth noting that the housing industry can also experience downturns and corrections. While exponential growth may occur for a period of time, factors such as economic recessions, rising interest rates, or oversupply of housing can lead to a downturn in the market (Grand View Research, 2021).

```python
# Test the normality assumption of the quantitative variables in the training dataset and
# define a function to test the normality of each column

def test_normality(column):
    p_value = stats.shapiro(column.fillna(0))[1]
    return p_value < 0.01

# Create a DataFrame with only the quantitative columns
quantitative = train.select_dtypes(include=['int64', 'float64']).columns.tolist()
quantitative_df = train[quantitative]

# Apply the normality test function to each column of the quantitative DataFrame
normality_results = quantitative_df.apply(test_normality)

# Print whether any of the columns failed the normality test
if not normality_results.any():
    print("All quantitative columns pass normality test with a significance level of 0.01.")
else:
    print("Some quantitative columns fail normality test with a significance level of 0.01. No normal distributed.")
```

## Feature processing.

### Data cleaning.

```python
# Drop sales column and concatenate train + test features
train_features = train.drop(['SalePrice'], axis=1)
test_features = test
features = pd.concat([train_features, test_features]).reset_index(drop=True)

features.shape
```

```python
# Cleaning and filling data that mean ausence of home features.
features['MSSubClass'] = features['MSSubClass'].apply(str)
features['YrSold'] = features['YrSold'].astype(str)
features['MoSold'] = features['MoSold'].astype(str)
features['Functional'] = features['Functional'].fillna('Typ') 
features['Electrical'] = features['Electrical'].fillna("SBrkr") 
features['KitchenQual'] = features['KitchenQual'].fillna("TA") 
features["PoolQC"] = features["PoolQC"].fillna("None")
features['Exterior1st'] = features['Exterior1st'].fillna(features['Exterior1st'].mode()[0]) 
features['Exterior2nd'] = features['Exterior2nd'].fillna(features['Exterior2nd'].mode()[0])
features['SaleType'] = features['SaleType'].fillna(features['SaleType'].mode()[0])
```

```python
for col in ('GarageYrBlt', 'GarageArea', 'GarageCars'):
    features[col] = features[col].fillna(0)

for col in ['GarageType', 'GarageFinish', 'GarageQual', 'GarageCond']:
    features[col] = features[col].fillna('None')

for col in ('BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2'):
    features[col] = features[col].fillna('None')

features['MSZoning'] = features.groupby('MSSubClass')['MSZoning'].transform(lambda x: x.fillna(x.mode()[0]))


features.head()
```

### Missing values treatment.

```python
# filling missing values for categorical features

objects = []
for a in features.columns:
    if features[a].dtype == object:
        objects.append(a)
features.update(features[objects].fillna('None'))

features['LotFrontage'] = features.groupby('Neighborhood')['LotFrontage'].transform(lambda x: x.fillna(x.median()))

numeric_dtypes = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
numerics = []
for a in features.columns:
    if features[a].dtype in numeric_dtypes:
        numerics.append(a)
features.update(features[numerics].fillna(0))
```

```python
# filling missing values for numerical features

numeric_dtypes = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
numerics2 = []
for a in features.columns:
    if features[a].dtype in numeric_dtypes:
        numerics2.append(a)
skew_features = features[numerics2].apply(lambda x: skew(x)).sort_values(ascending=False)

high_skew = skew_features[skew_features > 0.5]
skew_index = high_skew.index

for a in skew_index:
    features[a] = boxcox1p(features[a], boxcox_normmax(features[a] + 1))
```

### Feature grouping.

```python
# Grouping features by type

features = features.drop(['Utilities', 'Street', 'PoolQC',], axis=1)

features['YrBltAndRemod']=features['YearBuilt']+features['YearRemodAdd']
features['TotalSF']=features['TotalBsmtSF'] + features['1stFlrSF'] + features['2ndFlrSF']

features['Total_sqr_footage'] = (features['BsmtFinSF1'] + features['BsmtFinSF2'] +
                                 features['1stFlrSF'] + features['2ndFlrSF'])

features['Total_Bathrooms'] = (features['FullBath'] + (0.5 * features['HalfBath']) +
                               features['BsmtFullBath'] + (0.5 * features['BsmtHalfBath']))

features['Total_porch_sf'] = (features['OpenPorchSF'] + features['3SsnPorch'] +
                              features['EnclosedPorch'] + features['ScreenPorch'] +
                              features['WoodDeckSF'])
```

```python
# Cross-validation object of root mean squared error and logarithmic 

kfolds = KFold(n_splits=10, shuffle=True, random_state=42)

def rmsle(y, y_pred):
    return np.sqrt(mean_squared_error(y, y_pred))

def cv_rmse(model, X=X):
    rmse = np.sqrt(-cross_val_score(model, X, y, scoring="neg_mean_squared_error", cv=kfolds))
    return (rmse)
```

# Model Blending.

Blending different regression models can have several benefits, including:

1. **Improved accuracy:** since every model has its strengths and weaknesses, blending them together help to balance out these differences and produce more accurate results overall.

2. **Reduced overfitting:** overfitting occurs when a model is too complex and captures noise in the data instead of the underlying patterns. Blending models can reduce overfitting and improve the generalization of the model.

3. **Increases robustness:** errors on one model may be compensated be the strengths of others.

4. **Flexibility:** using linear regression model can capture linear relationships in the data, and a tree-based model like XGBoost or LightGBM can be used to capture more complex, non-linear relationships.

```python
# Blending models and assigning them a weight, based on RSME and estimation of the relevance of each model.

def blend_models_predict(X):
    return ((0.1 * elastic_model_full_data.predict(X)) + \
            (0.05 * lasso_model_full_data.predict(X)) + \
            (0.1 * ridge_model_full_data.predict(X)) + \
            (0.1 * svr_model_full_data.predict(X)) + \
            (0.1 * gbr_model_full_data.predict(X)) + \
            (0.15 * xgb_model_full_data.predict(X)) + \
            (0.1 * lgb_model_full_data.predict(X)) + \
            (0.3 * stack_gen_model.predict(np.array(X))))

print('RMSLE score on train data:')
print(rmsle(y, blend_models_predict(X)))
```
