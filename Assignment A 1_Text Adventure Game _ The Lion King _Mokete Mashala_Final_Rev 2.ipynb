```python
"""
Hult International Business School
@Date: Wednesday, 1 February 2023
@Author: Mokete Mashala
"""
"""
    DocString:
    
    A) Description: 
    A) Introduction:
    This game is based on the animated musical drama film "The Lion King" 
    set in the kindom of lions in Africa. It was produced by Walt Disney 
    Feature Animation in 1994. The game consists of four stages and user 
    defined functions for start, win and lose. The game requires some 
    knowledge of the movie, which will be helpful.
    
    The three rounds are specified as:

    Scene 1: Deception leads into exile
    Scene 2: Meal with friends 
    Scene 3: Falling in Love
    Scene 4: Combat: Defeat Scar and the hyenas
    
    B) Known Bugs and/or Errors:
    None at the time of the test.
    
    References :  https://en.wikipedia.org/wiki/The_Lion_King
               :  https://www.imdb.com/title/tt0110357/plotsummary
    
  """  
###########################################################################
############# Import packages that will be used in the game ###############
###########################################################################

# package to exit the game
from sys import exit   

# the package to choose an object randomly from a list
import random 

# the package to add the sleep function to pause allow more time between output to user
import time 

# the package to convert data into a dataframe
import pandas as pd

#import matplotlib
from matplotlib import pyplot as plt


############################################################################
#################### define introduction function ##########################
#
#    intro function introduces the game and provides the background 
#    of the game and the movie on which the game is based on.
#
############################################################################

def intro():
    """ 
    intro function introduces the game and provides the background 
    of the game and the movie on which the game is based on. 
    """
    
    print("")
    print("In this adventure, you are Simba, the lion hero ü¶Å.")
    print("")
    print("The Cirle of Life... ")
    print("In the Pride Lands of Africa...you have just been born.")
    print(f""" 
    You are set to succeed your father, 
    Mufasa as the King of the Pride Lands.
    The wise choices you make, the strength üí™ you have will 
    influence the outcome of the game in each stage.\n""")
    
    print("Welcome to the world... Your adventure awaits you")
    
    input("<Press enter to continue>\n")
    
    create_name()  
    
    
############################################################################
#################### define introduction function ##########################
############################################################################

def create_name():
    """
    create_name function gives the opportunity for the user to input their name
    and also confirm that a name has been provided. 
    """   
    
    print("*"*100)

    user_name = input(prompt = f"""    
    What is your name?
    >>>  """)
    
    #user name is only allowed to be characters, no digits or empty input allowed
    if user_name == "" or user_name.isdigit():
        print("NameError: you did not enter a proper name, try again!")
        create_name()
        
    else:
        print("*"*100)
        print(f"""
    Welcome to the game {user_name.strip().upper()}""")
        obey_or_not()

    
############################################################################
#################### define introduction function ##########################
#
# Obey_or_not function gives sets the scene for when Simba loses his father 
# upon being deceived by Scar.
# 
#
############################################################################

def obey_or_not():
   
    """ 
    obey_or_not function gives sets the scene for when Simba loses his father 
    upon being deceived by Scar.
    The player is given the chance to obey his father Mufasa and go into a 
    skill test or to disobey and be sent to the desert. Failure to pass the skill
    test will also result in player being sent to desert.
    Passing the skill test, will allow the player to skip into the next scene 
    without going into the desert.
    """
    
    print("*"*100)
      
    # this is the list of words or phrases that the user can input
    secret_phrase = ["hakuna mathata", "mathata", "hakuna", "no worries", \
                 "love", "deception", "obedience"]  
    
    #In the fist interaction with the variable choice it is unknown, therefore set
    # it to none.
    
    choice = None
    
    #while the choice does not exist as yet.
    while choice is None:
        print("""
        You are with your uncle Scar at the Pride Rock. 
        Your father Mufasa has warned you to stay away from him. 
        Your uncle Scar tells you that he has a surprise for you. 
        He tells you to follow him.
        What will you do?""")
   
        print(f"""
        1) Disobey your father and follow him
        2) Obey your father and go back home!""")
    
        choice = input(">>> \n")
        print("*"*100)    
    
    # choice 1 and control for other possible user input
    # go into exile
    
    if '1' in choice or 'disobey' in choice or 'follow' in choice:
        print("""
        You have decided to disobey your father. You now find yourself 
        surrounded by a large herd of wildebeest. You hang onto a tree. 
        In the midst of the stampede you see your father Mufasa. 
        He rescues you from the stampede, but unfortunately he does not survive. 
        Scar deceives you into believing that it is your fault so go into exile\n""")
        
        input("<Press enter to continue>\n")
        
        time.sleep(0.5)
    
        meal_time()

        print('\n')
            
    # choice 2 and control for other possible user input
    # in this option the user will be given a skill test based on knowledge of the movie
    
    elif '2' in choice or 'obey' in choice or 'back' in choice or 'home' in choice:
        print(f"""
        You made the right choice, You stand a good chance of being the next King!""")
                
        time.sleep(0.5)
        
        input("<Press enter to continue>\n")
        
        print("*"*100)
        print(f"""
        As you are walking home, you are proud of making the right decision.
        However, Scar is upset and plans more evil against you.""")
        
        print(f"""
        To make it safely home, you must provide a secret phrase or significant word.""")       
        
        time.sleep(0.5)
        user_input = input(prompt= f"""
        What is the moral lesson of the Lion King Movie.""")
        
        #the user is allowed three attempts in total, so attempt 3 is counting 
        # from the first given attempt before the countdown
        attempt = 2
        
        while True:
                       
            if user_input.lower() in secret_phrase:
                print("*"*100) 
                print("\nYou can safely make your way home!")
                
                input("<Press enter to continue>\n")
                love()
            
            elif attempt > 0 and user_input not in secret_phrase:
                print("*"*100)            
                user_input = input(prompt = f"""\n
        That's not the correct password. Please try again
        You have {attempt} attempts left. 
        What is the moral lesson of the Lion King Movie: 
        I will give you a hint: It means no worries! """)
                
            elif attempt == 0:
                print("*"*100) 
                print("""
        You did not make it! Hakuna Mathata, It means no worries!
        Evil is upon you!
        Scar murdered your father and told everyone it was your fault.
        He deceives you into believing that it is your fault so go into exile\n""")
                
                input("<Press enter to continue>\n")
                meal_time()  # call the next scene function which is meal_time
            
            attempt -= 1  # this will reduce the attempt after each round
            
        print('\n')
    
   #control for user input. User will be taken back to the start of the function/scene
    else:
        print("You did not input the correct answer. Please try again.")
        input("<Press enter to continue>\n")
        obey_or_not()  # call the next scene function which is obey_or_not
      
    print('\n')

    
############################################################################
#################### define introduction function ##########################
############################################################################

def meal_time():
    """ 
    meal_time function sets the scene where Simba is in the desert. He is saved
    by Timon and Pumbaa and they all go to the jungle of paradise.
    They show Simba the type of food they eat and he will need to choose 
    whether to eat the gross looking worms or not. If he does not eat, he loses
    the game because eventually he goes hungry and dies.
    """
    time.sleep(0.5)
    print("*"*100)
    character_strength = 20
   
    
    #at the start the scene the user is give a strength of 20 points
    #the user will be given a choice to wonder in the desert looking for food
    #if the user wonders too long then the user will lose the game and die
    
    
    print(f"""
     For this stage, your initial strength üí™ is {character_strength}""")
    
    choice = None
    
    while choice is None:
        time.sleep(0.5)
        print(f"""
        You are in the desertüåµ!
        You are hungry and there is no water!
        You collapse from hunger and thirst and vultures surround you!
        Suddenly the warthog, Pumbaa and meerkat Timon come to rescue you.
        
        You are now in the jungle paradise and are still hungry
        Your friends invite you to eat slimy bugs and worms with them. 
        What will you do?
        
        ü™±üêõüêû
        1) eat, because you are hungry.
        2) refuse to eat, it looks too gross.
        3) keep walking to find alternative food.
        """)
        time.sleep(0.5)
        choice = input(">>> \n")
        print("*"*100) 
    
    if choice == "1" or "eat" in choice or "hungry" in choice:
        print(f"""
        That was slimy, yet satisying.
        You eventually grow into a healty, carefree Lion ü¶Å.
        Your strength üí™ remains at {character_strength}.
        You may proceed. Strong and mighty Lion \n""")
        
        input("<Press enter to continue>\n")
        
        love()
        print('\n')
            
    elif choice =="2" or "refuse" in choice or "gross" in choice:
        print("""
        Because you refused to eat. You die of hunger in jungle of paradise!""")
        character_strength = 0
        
        print(f"""
        Your strength üí™ is now {character_strength}""")
        
        input("<Press enter to continue>\n")
        
        fail()  # fail function. It means that the user has lost the game
        exit(0)  #exit function. It means the user must exit the game
        
    elif choice =="3" or "keep" in choice or "walking" in choice or \
    "alternative" in choice:
        
        # if user chooses not to eat, another chance is give.
                print(f"""
        You have been walking around for three hours and have still found no food.
        Your energy has depleted even more!
        You need to eat something very soon or else you might die of hunger!
        
        What will you do?
        
        ü™±üêõüêû
        1) eat, because you are hungry.
        2) continue keep walking around to find alternative food.
        """)
                time.sleep(0.5)
                choice = input(">>> \n")
        
                if choice == "1" or "eat" in choice or "hungry" in choice:
                    print(f"""
        That was slimy, yet satisying.
        You eventually grow into a healty, carefree Lion ü¶Å.
        Your strength üí™ remains at {character_strength}.
        You may proceed. Strong and mighty Lion \n""")
        
                    input("<Press enter to continue>\n")
        
                    love()
                    print('\n')
        
                elif choice =="2" or "keep" in choice or "walking" in choice or \
                "alternative" in choice:
        
        # no more chances. You lose
                    print(f"""
        You have been walking around for too long now. You die of hunger!""")
            
                    input("<Press enter to continue>\n")
                    fail()  # fail function. It means that the user has lost the game
                    exit(0)  #exit function. It means the user must exit the game
  
    else:
        print("You did not input the correct answer. Please try again.")
        
        input("<Press enter to continue>\n")
        print(f"""
        You are now back to where you started, Please try again.""")
        meal_time()

        
############################################################################
#################### define introduction function ##########################
############################################################################

def love():
    """
    love function defines the love scene. It is based on the scene where
    Simba is with Nala in the jungle of paradise and the two of them are 
    having a conversation. Later in the scene there is a romantic scene where
    there is love music in the background and the two characters fall in love.
    Based on a roll of a die the user will have a chance to fall in love with Nala
    """
    
    choice = None
    
    while choice is None:
        time.sleep(0.5)
        print("*"*100)
        print(f"""
        Your friend Nala found you in the jungle of paradise.
        You have been playing and having fun the whole day and now its nighttime
        The sky is full of stars...And the moon is shining bright.
        The birds are singing üé∂!!!
        There is a stream of water flowing by !!!""")
        
        time.sleep(0.5)
        print(f"""
        Can you feel the love tonight üíïüíïüíï? 
        1) yes!
        2) no!
        """)
        
        choice = input(">>>  \n")
        print("*"*100)
    
        if "1" in choice or "yes" in choice:
            print(f"""
            Ooohhh that is so sweet!""")
              
            # Initiate a dictionary with empty list to store results for the 
            # die rolling. "Is there love" key will compare results of rolling 
            # the die at each round "yes" occurs when both roll same number 
            # while "no" occurs if they don't
            
            total_roll = {'Rounds':[], 'Simba':[],'Nala': [], 'Is there love':[] }   
            is_there_love = [] #empty list to store the results for matching die rolls
                                                         
            print(f"""
            You and Nala will roll a die 7 times to see if she will fall in love with you!
            If she falls in love with you...It will be forever...""")
            input("<Press enter to roll the die>")

            simba_dice = random.randint(1,6)
            nala_dice = random.randint(1,6)
            
            # a for loop to allow the user and Nala to roll the die
            
            for i in range(1,8):
                simba_dice = random.randint(1,6)
                nala_dice = random.randint(1,6)
                total_roll['Rounds'].append(i)  #append the outcome into the dictionary
                total_roll['Simba'].append(simba_dice) #append into the dictionary
                total_roll['Nala'].append(nala_dice)   #append into the dictionary

            # declare a roll_index variable as zero, this will increase each time 
            # the index is used to indicate at which point a "yes" (i.e. when
            # both players had the same number rolled by a die) appeared in the 
            #7 rounds of rolling the die.
            roll_index = 0
            
            #for loop to append the dictionary with the results of "yes" or "no"
            #depending on the roll of the die being same or not
            
            for roll_simba in total_roll['Simba']:
                if total_roll['Simba'][roll_index] == total_roll['Nala'][roll_index]:
                    love ="yes"
                    total_roll['Is there love'].append(love)
                elif total_roll['Simba'][roll_index] != total_roll['Nala'][roll_index]:
                    love = "no"
                    total_roll['Is there love'].append(love)
                is_there_love.append(love)
                roll_index += 1
                        
            df = pd.DataFrame(total_roll)  #convert the dictionary into a DataFrame
            df = df.set_index(['Rounds'])
            

            if "yes" in total_roll['Is there love']:
                time.sleep(0.5)
                print("*"*100)
                print(f"""
        You and Nala rolled the die {roll_index} times 
        Love is there üíïüíïüíï! 
        You both managed to roll the same number in round {is_there_love.index('yes')+1}!
        Congratulations, Your love will last forever!
        These were the results:""")
                
                #display the dataframe
                display(df) 
                
                input("<Press enter to continue>") 
               
                combat()
        
            else:
                time.sleep(0.5)
                print("*"*100)
                print (f"""
            You and Nala rolled the die {roll_index} times 
            Unfortunately, you did not roll any similar number in all rounds.
            You will remain without love for now! üíî.
            These were the results:""")
                display(df)
               
                input("<Press enter to continue>\n")
                combat()
    
        elif "2" in choice or "no" in choice:
            print("""\nYou are staring love in the face and you chose No! 
            That makes no sense!""")
            print("""\nYou will remain without love for now üíî.""")
               
            input("<Press enter to continue>\n")
            combat()
        
        else:
            print("You did not input the correct answer. Please try again.")
            input("<Press enter to continue>\n")
            exit(0)      

            
############################################################################
#################### define introduction function ##########################
############################################################################

def combat():
    print("*"*100)
        
    """This function describes the last scene of the game where Simba will go
    into battle with Scar
    At the beginning of the game they each have a set of strength. 
    Depending on the rolling die, their strength will get depleted. 
    The one who get to zero strength first will lose, whilst the opponent wins
    """
    
    character_strength = 20
    print(f"""
    Your strength üí™ is {character_strength}""")
    
    enriched_strength = None
    scar_strength = 30
    choice = None
    while choice is None:

        print (f"""
        You are in the open field.
        Rafiki is following you and reveals to you that he knows your father.
        He tells you that Scar has taken over the and everything has fallend apart.
        He tells you that your father lives in you and that:
        "You must go back home and take your rightful place as the king"
    
        What will you do?    
        1) go back home and fight for your kingdom.
        2) stay in the jungle of paradise.
        3) let your friends decide for you. """)
    
        choice = input(">>> \n")
              
        if "1" in choice or "back" in choice or "home" in choice or "fight" in choice:
            print("*"*100)
            print(f"""
            You have crossed the desert and arrived in the Pride Lands
            You confront Scar and demand he steps away.
            He refuses and yellsüò° :
            Why must everything end in violence?""")
            
            input("<Press enter to continue>\n")
            print("*"*100)
            
            scar_strength = 30
            
            print(f"Your current strength is {character_strength}")
            print(f"Scar's strength is {scar_strength}")
            
            print("You have final opportunity to increase your strength")
            user_input = input(prompt = "Enter a number between 1 and 10: ")
            
            #to control for user input convert string to integer
            try:
                user_input = int(user_input)
             
            #to control for user input if not interger then user to go back 
            #to start of combat
            except:
                print("You did not input an interger as specified, try again")
                input("<Press enter return to the combat starting point>\n")
                combat()              
             
            # if user input is less than ten, then the user to go back to 
            #start of combat
            if user_input > 10:
                print("Your number is larger than 10, try again")
                input("<Press enter return to the combat starting point>\n")
                combat() 
 
            # if user input is less than zero, then the user to go back to 
            #start of combat
            elif user_input < 0:
                print("Your number is negative, try again")
                input("<Press enter return to the combat starting point>\n")
                combat() 

            # if user input is more than 0 but less than 5, multiply user input
            # by 2
            elif user_input <= 0 and user_input <=5:
                enriched_strength = user_input * 2
                character_strength = character_strength + enriched_strength

            # if user input is more than 5 but less than and equal to 10, 
            # multiply user input by 1
            elif user_input > 5 and user_input <= 10:
                enriched_strength = int(user_input) * 1
                character_strength = character_strength + enriched_strength
                           
            print("*"*100)
            print(f"Your enriched strength is {character_strength}")
            print(f"Scar's strength is {scar_strength}")
            
            print(f"""
            The battle is about to start
            At the Pride Rock surrounded with flames you fight with Scar""")                 
                    
            input("<Press enter to continue>\n")
            print("*"*100)
                        
            # Below is the dictonary containing the results of each round of 
            #the battle
            combat_roll={"You": [], "Scar":[], "Outcome":[], "Your Strength":[],\
                         "Scar Strength":[]}

            #while loop is for when both of them still have strength greater than 
            #zero. If the strength is less than or equal to zero, the while loop
            # will stop.
            while scar_strength > 0 and character_strength > 0:  
                user_roll = random.randint(1,6)  #random interger
                
                #append the results of Roll to dictionary
                combat_roll["You"].append(user_roll)  
                
                scar_roll = random.randint(1,6)  #random interger
                
                #append the results of Roll to dictionary
                combat_roll["Scar"].append(scar_roll) 
                
                #if the user has a higher roll of a die, then the user hits Scar
                # the one who gets hit loses strength, so in this case Scar 
                #(opponent) will lose points
                if user_roll >= scar_roll:
                    result = "You hit Scar!"
                    combat_roll["Outcome"].append(result)
                    combat_roll["Your Strength"].append(character_strength)
                    combat_roll["Scar Strength"].append(scar_strength)
                    
                    #control the while loop from running infinitely
                    scar_strength -= scar_roll  
                
                # if the user is lower than scar then the user will lose strength
                else:
                    result = "Scar hits you!"
                    combat_roll["Outcome"].append(result)
                    combat_roll["Your Strength"].append(character_strength)
                    combat_roll["Scar Strength"].append(scar_strength)
                    
                    #control the while loop from running infinitely
                    character_strength -= user_roll                
                           
                # The dataframe contains the information about each stage of fight
                df_combat = pd.DataFrame(combat_roll)                
                
            # if the character strength which is the user strength is bigger that 
            # the opponent, then the user has won the game.
            
            if character_strength > scar_strength:
                
                #display the dataframe containing the outcome of each hit
                display(df_combat) 
                
                print("") #print empty line to create space
                
                
                #A plot of the strength of each player for each round
                #set the x and y values for plotting the outcome of each hit
                x = range(1,len(combat_roll["Outcome"])+1)
                y1 = combat_roll["Your Strength"]
                y2 = combat_roll["Scar Strength"]               
                
                # Ploting 
                plt.plot(x, y1, label="You", color = 'red', marker='o', \
                         linestyle='dashed')
                plt.plot(x, y2, label = "Scar", color = 'blue', marker='*')
                
                #Add title
                plt.title("Fight for the Kingship")
                
                #Add legend
                plt.legend(bbox_to_anchor=(1,1), loc="upper left")
                
                #display plot
                plt.show()    
                    
                #Call the function of new_king() for victory by Simba (i.e. user)
                new_king()
            
            #if the charater strength is less then scar will win the game
            elif character_strength < scar_strength:
                
                #display the dataframe containing the outcome of each hit
                display(df_combat) 
                
                print("")#print empty line to create space
                
                #A plot of the strength of each player for each round
                #set the x and y values for plotting the outcome of each hit
                x = range(1,len(combat_roll["Outcome"])+1)
                y1 = combat_roll["Your Strength"]
                y2 = combat_roll["Scar Strength"]               
                
                # Ploting 
                plt.plot(x, y1, label="You", color = 'red', marker='o', linestyle='dashed')
                plt.plot(x, y2, label = "Scar", color = 'blue', marker='*')
                
                #Add title
                plt.title("Fight for the Kingship")
                
                #Add legend
                plt.legend(bbox_to_anchor=(1,1), loc="upper left")
                
                #display plot
                plt.show()  
                
                #Call the function of fail() for loss by Simba (i.e. user)
                fail()  # fail function. It means that the user has lost the game
                
            #if it is ever a tie, the combat must restart 
            elif character_strength == scar_strength:  
                combat() # combat function. User allowed to restart
                
                
        #if user chooses 2, then the user losses the game      
        elif "2" in choice or "stay" in choice or "jungle" in choice\
                                or "paradise" in choice:
            print("""\nLooks like you are afraid of taking responsibility.""")
            character_strength = 0
            print(f"""
            Your strength üí™ is now {character_strength}""")
            fail() # fail function. It means that the user has lost the game
    
        # if user chooses to listen to the friends, the friends dont want to 
        #go to the battle to fight, so the user will lose.
        
        elif "3" in choice or "friends" in choice:
            print("\n Your friends prefer you stay in the jungle of paradase!")
            character_strength = 0
            print(f"""
            Your strength üí™ is now {character_strength}""")
            fail() # fail function. It means that the user has lost the game
        
        # to control the user input, else statmement to ask the user to start again.
        else:
            print("Invalid command. Please try again.")
            input("<Press enter to go back to combat>\n")
            
            # invalid input gives user another chance to go back and restart.
            combat() 


############################################################################
#################### define introduction function ##########################
#
#If the user wins the combat / fight with Scar, the user gets crown new king.
#
############################################################################

def new_king():
    """new_king function defines Simba as the winner of the combat 
    with Scar and is crowned King of the the Pride Rock"""
    print("*"*100)
    print("""
    You have fought the good fightüèÜ. You are now the King of Pride Rock üëë...""")
    print("")
    time.sleep(0.3)
    input("<Press enter to exit, you King of the Jungle üôåüôåüôå!>\n")
    exit(0)

    
############################################################################
#################### define introduction function ##########################
#
# If the user fails to win the combat / fight with Scar, the user loses.
#
############################################################################
    
def fail():
    """fail function defines the defeat of Simba in the combat.
    Simba loses the battel against Scar and fails to inherit the kingdom."""
    print("*"*100)
    print(f"""
    You have failed in your quest to inherit your kingdom üò•.""")
    print("")
    time.sleep(0.3)
    input("<Game Over, press enter to exit>\n")
    exit(0)    

    
############################################################################
#####################  starting the game  ##################################
############################################################################   
    
intro()    # This will begin the game
```
